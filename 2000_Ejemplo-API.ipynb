```python
import requests
import json
import pandas as pd
from matplotlib import pyplot as plt
```

```python
response = requests.get('https://api.exchange.ripio.com/api/v1/orderbook/btc_usdc')
api_response = response.json()
```

```python
# api_response
```

```python
api_response.keys()
```

```python
sell_df = pd.DataFrame(api_response['sell']).astype('float')
buy_df = pd.DataFrame(api_response['buy']).astype('float')
```

```python
plt.figure(figsize=(30, 5))
plt.hist(sell_df['price'], bins=20)
plt.hist(buy_df['price'], bins=20, alpha=0.5)
_=plt.xticks(rotation=90)
```

```python
sell_df['price'].hist(bins=20, figsize=(30,5))
buy_df['price'].hist(bins=20,)
```
